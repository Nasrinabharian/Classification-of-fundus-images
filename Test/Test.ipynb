```python
def isnotsame(x):
    if x[0]=='[':
        x=x[1:]
    l=len(x)
    if x[l-1]==']':
        x=x[0:l-1]
    if x=='0.':
        x='0'
    if x=='1.':
        x='1'
    return x
```

```python
def max1(x,y):
    s=float(x)+float(y)
    s=s/2
    return s
```
